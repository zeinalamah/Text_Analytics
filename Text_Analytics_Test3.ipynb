```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import requests
import PyPDF2
from PyPDF2 import PdfFileReader
import os
```

```python
# Get the list of all files and directories
# in the root directory
path = "/Text_Analytics_Data"
PDFs_list = os.listdir(path)
path_list=[]
pdf_files_objects=[]
pdf_reader_objects=[]

# Generate a list of paths for required files
for PDF in PDFs_list: 
    path_list.append(path+"/"+ PDF)


# creating a pdf files objects
for file in path_list: 
    pdf_files_objects.append(open(file, 'rb')) 

# creating a pdf reader objects
for f in range(0,len(pdf_files_objects)): 
    pdf_reader_objects.append(PyPDF2.PdfFileReader(pdf_files_objects[f]))
    

text = ""
text_list=[]

for book in range(0,len(pdf_files_objects)):
    for i in range(0,pdf_reader_objects[book].numPages):
        # creating a page object
        page_object = pdf_reader_objects[book].getPage(i)
        # extracting text from page
        text=text+page_object.extractText()
    text_list.append(text)
    text=''
    
    
#Closing the files we opened    
for file in range (0,len(pdf_files_objects)): 
    pdf_files_objects[file].close()

print(pdf_files_objects[1])
print(text_list[1])
```

```python
books_df=pd.DataFrame(text_list)
books_df.head()
df_columns=["Book","Content"]
dicts=[]

#Generating a list of dictionaries for the dataframe
for book in range (0,len(PDFs_list)):
    dicts.append(dict(zip(df_columns,[PDFs_list[book],text_list[book]])))

books_df=pd.DataFrame(dicts)


vectorizer = CountVectorizer()

# Use the content column instead of our single text variable
matrix = vectorizer.fit_transform(books_df.Content)
counts = pd.DataFrame(matrix.toarray(),
                  index=books_df.Book,
                  columns=vectorizer.get_feature_names())

counts.head()
```
